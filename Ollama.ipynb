## Importing Libraries and ambient variables

```python
import os
import threading
import requests
import subprocess
import requests
import json

def ollama():
    os.environ["OLLAMA_HOST"] = "0.0.0.0:11434"
    os.environ["OLLAMA_ORIGINS"]= '*'
    subprocess.Popen(['ollama', "serve"])


ollama()
```

```python
# Once Ollama is installed, you may use it in your Colab as it follows:
# Start the Ollama server executing the following code:
```

```python
ollama_thread = threading.Thread(target=ollama)
ollama_thread.start()
```

## Prompt

```python
prompt = """
What is Artificial Intelligence?
"""
```

## Output

```python
url = 'http://localhost:11434/api/chat'
payload = {
    "model": "llama3.2",  # Make sure to use the correct name
    "temperature": 0.6,
    "stream": False,
    "messages": [
        {"role": "system", "content": "You are AI assistant"},
        {"role": "user", "content": prompt}
    ]
}
response = requests.post(url, json=payload)
message_str = response.content.decode('utf-8')
message_dict = json.loads(message_str)
print(message_dict)
```
